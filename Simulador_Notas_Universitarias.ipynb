```python
# Importamos las herramientas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuracin esttica de los grficos
sns.set_theme(style="whitegrid")

# 1. SIMULACIN: Generar datos aleatorios
# Vamos a simular las notas de 1000 estudiantes.
# Suponemos una media (mu) de 7 y una desviacin estndar (sigma) de 1.5
mu = 7
sigma = 1.5
datos_simulados = np.random.normal(loc=mu, scale=sigma, size=1000)

# 2. VISUALIZACIN (EDA - Exploratory Data Analysis)
plt.figure(figsize=(10, 6))

# Creamos un histograma (barras) y la curva de densidad (lnea suave)
sns.histplot(datos_simulados, kde=True, color="blue", alpha=0.6)

# Agregamos lneas verticales para la media terica
plt.axvline(x=mu, color='red', linestyle='--', label=f'Media Terica ({mu})')
plt.axvline(x=np.mean(datos_simulados), color='green', linestyle='-', label=f'Media Real Simulada ({np.round(np.mean(datos_simulados), 2)})')

plt.title('Simulacin de Notas Universitarias (Distribucin Normal)', fontsize=16)
plt.xlabel('Nota (0-10)')
plt.ylabel('Frecuencia (Cantidad de Estudiantes)')
plt.legend()
plt.show()
```
