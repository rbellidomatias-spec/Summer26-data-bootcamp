```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Estética de los gráficos
sns.set_theme(style="whitegrid")

# 1. GENERAMOS LOS DATOS
# Imaginemos una Pyme: 10 empleados ganan normal, 1 dueño gana una fortuna.
sueldos = [2500, 2600, 2400, 2550, 2700, 2800, 2500, 2650, 2750, 2450]
sueldos.append(3000) # <--- EL OUTLIER (El dato que rompe todo)

# Creamos la tabla (DataFrame)
df = pd.DataFrame(sueldos, columns=['Sueldo_USD'])

# 2. CALCULAMOS ESTADÍSTICOS
promedio = df['Sueldo_USD'].mean()
mediana = df['Sueldo_USD'].median()

print(f"--- RESULTADOS ---")
print(f"El Promedio (Media) dice que se gana:  ${promedio:,.2f}")
print(f"La Mediana (Realidad) dice que se gana: ${mediana:,.2f}")

# 3. GRAFICAMOS (Boxplot)
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['Sueldo_USD'], color="tomato")
plt.title('Detección de Outliers: El punto aislado a la derecha', fontsize=12)
plt.xlabel('Sueldo en USD')
plt.show()

# 4. MEDIDAS DE DISPERSIÓN (El Riesgo)

# Grupo A: Todos ganan parecido (Poca variabilidad)
grupo_a = [2500, 2550, 2450, 2600, 2500]

# Grupo B: Unos ganan muy poco y otros mucho (Alta variabilidad)
grupo_b = [1000, 4000, 1200, 3800, 2600]

# Convertimos a DataFrames
df_a = pd.DataFrame(grupo_a, columns=['Sueldo'])
df_b = pd.DataFrame(grupo_b, columns=['Sueldo'])

print(f"Promedio Grupo A: ${df_a['Sueldo'].mean():.2f}")
print(f"Promedio Grupo B: ${df_b['Sueldo'].mean():.2f}")
print("-" * 30)
print(f"Desviación Estándar Grupo A: ${df_a['Sueldo'].std():.2f}")
print(f"Desviación Estándar Grupo B: ${df_b['Sueldo'].std():.2f}")

# Gráfico comparativo
plt.figure(figsize=(10, 4))
# Juntamos los datos para graficar
datos_juntos = pd.DataFrame({
    'Grupo A': grupo_a,
    'Grupo B': grupo_b
})
sns.boxplot(data=datos_juntos)
plt.title('Mismo Promedio, Distinta Dispersión (Riesgo)', fontsize=14)
plt.show()
```
